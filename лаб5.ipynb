```python
# 1 задание

class Vertex:
    def __init__(self, key):
        self.id = key
        self.connected = {}

    def addNeighbor(self, nbr, weight=0):
        self.connected[nbr] = weight

    def __str__(self):
        return str(self.id) + ' connected: ' + str([x.id for x in self.connected])

    def getConnections(self):
        return self.connected.keys()

    def getId(self):
        return self.id

    def getWeight(self, nbr):
        return self.connected[nbr]


class Graph:
    def __init__(self):
        self.vertList = {}
        self.numVertices = 0

    def addVertex(self, key):
        self.numVertices += 1
        newVertex = Vertex(key)
        self.vertList[key] = newVertex
        return newVertex

    def getVertex(self, n):
        return self.vertList.get(n)

    def __contains__(self, n):
        return n in self.vertList

    def addEdge(self, f, t, weight=0):
        if f not in self.vertList:
            self.addVertex(f)
        if t not in self.vertList:
            self.addVertex(t)
        self.vertList[f].addNeighbor(self.vertList[t], weight)

    def getVertices(self):
        return self.vertList.keys()

    def __iter__(self):
        return iter(self.vertList.values())


# Поиск в ширину (BFS)
from collections import deque

def bfs(graph, start):
    start_vert = graph.getVertex(start)
    if not start_vert:
        return []

    visited = set()
    queue = deque([start_vert])
    order = []

    while queue:
        vert = queue.popleft()
        if vert.getId() not in visited:
            visited.add(vert.getId())
            order.append(vert.getId())
            for nbr in vert.getConnections():
                if nbr.getId() not in visited:
                    queue.append(nbr)
    return order


if __name__ == "__main__":
    g = Graph()
    g.addEdge('A', 'B')
    g.addEdge('A', 'C')
    g.addEdge('B', 'D')
    g.addEdge('C', 'D')
    g.addEdge('D', 'E')

    print("BFS обход от A:", bfs(g, 'A'))
```

```python
# 2 задание

class Vertex:
    def __init__(self, key):
        self.id = key
        self.connected = {}

    def addNeighbor(self, nbr, weight=0):
        self.connected[nbr] = weight

    def getConnections(self):
        return self.connected.keys()

    def getId(self):
        return self.id


class Graph:
    def __init__(self):
        self.vertList = {}

    def addVertex(self, key):
        newVertex = Vertex(key)
        self.vertList[key] = newVertex
        return newVertex

    def getVertex(self, n):
        return self.vertList.get(n)

    def addEdge(self, f, t, weight=0):
        if f not in self.vertList:
            self.addVertex(f)
        if t not in self.vertList:
            self.addVertex(t)
        self.vertList[f].addNeighbor(self.vertList[t], weight)

    def getVertices(self):
        return self.vertList.keys()


def reverse_graph(graph):
    reversed_g = Graph()
    for v in graph.getVertices():
        for nbr in graph.getVertex(v).getConnections():
            reversed_g.addEdge(nbr.getId(), v)
    return reversed_g


def print_graph(graph):
    for v in graph.getVertices():
        neighbors = [nbr.getId() for nbr in graph.getVertex(v).getConnections()]
        print(f"{v} -> {neighbors}")

if __name__ == "__main__":
    g = Graph()
    g.addEdge('A', 'B')
    g.addEdge('A', 'C')
    g.addEdge('B', 'D')
    g.addEdge('C', 'D')
    g.addEdge('D', 'E')

    print("исходный граф")
    print_graph(g)

    rev = reverse_graph(g)
    print("\nобращённый граф")
    print_graph(rev)
```

```python
# 3 задание

class Vertex:
    def __init__(self, key):
        self.id = key
        self.connected = {}
        self.visited = False

    def addNeighbor(self, nbr, weight=0):
        self.connected[nbr] = weight

    def getConnections(self):
        return self.connected.keys()

    def getId(self):
        return self.id


class Graph:
    def __init__(self):
        self.vertList = {}

    def addVertex(self, key):
        newVertex = Vertex(key)
        self.vertList[key] = newVertex
        return newVertex

    def getVertex(self, n):
        return self.vertList.get(n)

    def addEdge(self, f, t, weight=0):
        if f not in self.vertList:
            self.addVertex(f)
        if t not in self.vertList:
            self.addVertex(t)
        self.vertList[f].addNeighbor(self.vertList[t], weight)

    def getVertices(self):
        return self.vertList.keys()


def dfs_topological(graph, vertex, stack):
    vertex.visited = True
    for nbr in vertex.getConnections():
        if not nbr.visited:
            dfs_topological(graph, nbr, stack)
    stack.append(vertex.getId())


def topological_sort(graph):
    stack = []
    for v in graph.getVertices():
        vertex = graph.getVertex(v)
        if not vertex.visited:
            dfs_topological(graph, vertex, stack)
    return stack[::-1]


# граф рецепта блинов
if __name__ == "__main__":
    g = Graph()
    # зависимости A -> B означает "A должно быть сделано до B"
    g.addEdge('разогреть сковороду', 'вылить смесь')
    g.addEdge('смешать ингредиенты', 'вылить смесь')
    g.addEdge('вылить смесь', 'перевернуть блин')
    g.addEdge('перевернуть блин', 'дожарить')
    g.addEdge('дожарить', 'полить сиропом')
    g.addEdge('полить сиропом', 'есть')

    order = topological_sort(g)
    print("порядок действий")
    for i, step in enumerate(order, 1):
        print(f"{i}. {step}")
```

```python
# 4 задание

import heapq

class Vertex:
    def __init__(self, key):
        self.id = key
        self.connected = {}
        self.dist = float('inf')
        self.prev = None

    def addNeighbor(self, nbr, weight=0):
        self.connected[nbr] = weight

    def getConnections(self):
        return self.connected.keys()

    def getId(self):
        return self.id

    def getWeight(self, nbr):
        return self.connected.get(nbr, float('inf'))


class Graph:
    def __init__(self):
        self.vertList = {}

    def addVertex(self, key):
        newVertex = Vertex(key)
        self.vertList[key] = newVertex
        return newVertex

    def getVertex(self, n):
        return self.vertList.get(n)

    def addEdge(self, f, t, weight=0):
        if f not in self.vertList:
            self.addVertex(f)
        if t not in self.vertList:
            self.addVertex(t)
        self.vertList[f].addNeighbor(self.vertList[t], weight)

    def getVertices(self):
        return self.vertList.keys()


def dijkstra(graph, start, end):
    start_vert = graph.getVertex(start)
    end_vert = graph.getVertex(end)

    if not start_vert or not end_vert:
        return None, None

    for v in graph.getVertices():
        graph.getVertex(v).dist = float('inf')
        graph.getVertex(v).prev = None

    start_vert.dist = 0
    pq = [(0, start_vert)]

    while pq:
        curr_dist, curr_vert = heapq.heappop(pq)

        if curr_dist > curr_vert.dist:
            continue

        if curr_vert == end_vert:
            break

        for nbr in curr_vert.getConnections():
            new_dist = curr_vert.dist + curr_vert.getWeight(nbr)
            if new_dist < nbr.dist:
                nbr.dist = new_dist
                nbr.prev = curr_vert
                heapq.heappush(pq, (new_dist, nbr))

    if end_vert.dist == float('inf'):
        return None, None

    path = []
    curr = end_vert
    while curr:
        path.append(curr.getId())
        curr = curr.prev
    path.reverse()

    return end_vert.dist, path


# создание графа
def create_graph():
    g = Graph()

    g.addEdge('Белово', 'Ленинск-Кузнецкий', 40)
    g.addEdge('Белово', 'Гурьевск', 30)

    g.addEdge('Ленинск-Кузнецкий', 'Гурьевск', 50)
    g.addEdge('Ленинск-Кузнецкий', 'Мыски', 120)
    g.addEdge('Ленинск-Кузнецкий', 'Кемерово', 100)
    g.addEdge('Ленинск-Кузнецкий', 'Новосибирск', 200)

    g.addEdge('Киселевск', 'Ленинск-Кузнецкий', 60)
    g.addEdge('Киселевск', 'Новосибирск', 250)

    g.addEdge('Гурьевск', 'Мыски', 90)

    g.addEdge('Мыски', 'Кемерово', 80)

    g.addEdge('Новосибирск', 'Кемерово', 210)

    return g


if __name__ == "__main__":
    g = create_graph()

    print("доступные города ", list(g.getVertices()))
    print()

    start = input("город отправления: ")
    end = input("город назначения: ")

    dist, path = dijkstra(g, start, end)

    if dist is None:
        print(f"\nпути из {start} в {end} не существует")
    else:
        print(f"\nкратчайшее расстояние: {dist} км")
        print(f"Путь: {' -> '.join(path)}")
```
