```python
# задание 1

class BinaryTree:
    def __init__(self, root):
        self.key = root
        self.left = None
        self.right = None

    def insertLeft(self, newNode):
        if self.left is None:
            self.left = BinaryTree(newNode)
        else:
            t = BinaryTree(newNode)
            t.left = self.left
            self.left = t

    def insertRight(self, newNode):
        if self.right is None:
            self.right = BinaryTree(newNode)
        else:
            t = BinaryTree(newNode)
            t.right = self.right
            self.right = t

    def getLeft(self):
        return self.left

    def getRight(self):
        return self.right

    def setRootVal(self, obj):
        self.key = obj

    def getRootVal(self):
        return self.key


def build_parse_tree(exp):
    tokens = exp.split()
    stack = []
    tree = BinaryTree('')
    stack.append(tree)
    current = tree

    for token in tokens:
        if token == '(':
            current.insertLeft('')
            stack.append(current)
            current = current.getLeft()
        elif token == 'не':
            current.setRootVal(token)
            current.insertRight('')
            stack.append(current)
            current = current.getRight()
        elif token in ('и', 'или'):
            current.setRootVal(token)
            current.insertRight('')
            stack.append(current)
            current = current.getRight()
        elif token == ')':
            current = stack.pop()
        else:  # операнд (истина/ложь)
            current.setRootVal(token == 'истина')
            current = stack.pop()
    return tree


def evaluate(tree):
    if tree.getLeft() is None and tree.getRight() is None:
        return tree.getRootVal()
    else:
        op = tree.getRootVal()
        if op == 'не':
            return not evaluate(tree.getRight())
        elif op == 'и':
            return evaluate(tree.getLeft()) and evaluate(tree.getRight())
        elif op == 'или':
            return evaluate(tree.getLeft()) or evaluate(tree.getRight())


def printexp(tree):
    if tree.getLeft() is None and tree.getRight() is None:
        return str(tree.getRootVal()).lower()
    else:
        op = tree.getRootVal()
        if op == 'не':
            return f"(не {printexp(tree.getRight())})"
        else:
            return f"({printexp(tree.getLeft())} {op} {printexp(tree.getRight())})"


# Тест
if __name__ == "__main__":
    exp = "( истина и ложь )"
    tree = build_parse_tree(exp)
    print("выражение ", printexp(tree))
    print("результат ", evaluate(tree))
```
