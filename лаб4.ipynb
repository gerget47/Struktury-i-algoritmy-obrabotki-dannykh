```python
# 1 задание

class HashTable:
    def __init__(self, size=11):
        self.size = self._next_prime(size)
        self.slots = [None] * self.size
        self.data = [None] * self.size
        self.count = 0

    def _next_prime(self, n):
        p = n if n % 2 else n + 1
        while True:
            if all(p % i != 0 for i in range(3, int(p**0.5) + 1, 2)):
                return p
            p += 2

    def _hash(self, key, step=0):
        return (hash(key) % self.size + step * step) % self.size

    def put(self, key, val):
        if self.count / self.size >= 0.7:
            self._resize(self.size * 2)
        step = 0
        start = self._hash(key)
        pos = start
        while self.slots[pos] not in (None, key):
            step += 1
            pos = self._hash(key, step)
        if self.slots[pos] is None:
            self.count += 1
        self.slots[pos] = key
        self.data[pos] = val

    def get(self, key):
        step = 0
        start = self._hash(key)
        pos = start
        while self.slots[pos] is not None:
            if self.slots[pos] == key:
                return self.data[pos]
            step += 1
            pos = self._hash(key, step)
        return None

    def __getitem__(self, key):
        return self.get(key)

    def __setitem__(self, key, val):
        self.put(key, val)

    def __len__(self):
        return self.count

    def __contains__(self, key):
        return self.get(key) is not None

    def __delitem__(self, key):
        step = 0
        start = self._hash(key)
        pos = start
        while self.slots[pos] is not None:
            if self.slots[pos] == key:
                self.slots[pos] = None
                self.data[pos] = None
                self.count -= 1
                if self.count / self.size <= 0.2 and self.size > 3:
                    self._resize(self.size // 2)
                return
            step += 1
            pos = self._hash(key, step)
        raise KeyError(f"Ключ {key} не найден")

    def _resize(self, new_size):
        old_slots, old_data = self.slots[:], self.data[:]
        self.size = self._next_prime(new_size)
        self.slots = [None] * self.size
        self.data = [None] * self.size
        self.count = 0
        for k, v in zip(old_slots, old_data):
            if k is not None:
                self.put(k, v)


if __name__ == "__main__":
    ht = HashTable()
    for i in range(20):
        ht[i] = i * 10
    print("Размер таблицы:", len(ht))
    print("10 в таблице?", 10 in ht)
    print("Значение по ключу 5:", ht[5])
    del ht[5]
    print("5 в таблице?", 5 in ht)
    print("Размер после удаления:", len(ht))
```

```python
# 2 задание

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class UnorderedList:
    def __init__(self):
        self.head = None

    def isEmpty(self):
        return self.head is None

    def add(self, item):
        temp = Node(item)
        temp.next = self.head
        self.head = temp


class HashTableChains:
    def __init__(self, size=11):
        self.size = self._next_prime(size)
        self.table = [UnorderedList() for _ in range(self.size)]
        self.count = 0

    def _next_prime(self, n):
        p = n if n % 2 else n + 1
        while True:
            if all(p % i != 0 for i in range(3, int(p**0.5) + 1, 2)):
                return p
            p += 2

    def _hash(self, key):
        return hash(key) % self.size

    def put(self, key, val):
        if self.count / self.size >= 0.7:
            self._resize(self.size * 2)
        index = self._hash(key)
        bucket = self.table[index]
        current = bucket.head
        while current:
            if current.data[0] == key:
                current.data = (key, val)
                return
            current = current.next
        bucket.add((key, val))
        self.count += 1

    def get(self, key):
        index = self._hash(key)
        bucket = self.table[index]
        current = bucket.head
        while current:
            if current.data[0] == key:
                return current.data[1]
            current = current.next
        return None

    def __getitem__(self, key):
        return self.get(key)

    def __setitem__(self, key, val):
        self.put(key, val)

    def __len__(self):
        return self.count

    def __contains__(self, key):
        return self.get(key) is not None

    def __delitem__(self, key):
        index = self._hash(key)
        bucket = self.table[index]
        current = bucket.head
        prev = None
        while current:
            if current.data[0] == key:
                if prev:
                    prev.next = current.next
                else:
                    bucket.head = current.next
                self.count -= 1
                if self.count / self.size <= 0.2 and self.size > 3:
                    self._resize(self.size // 2)
                return
            prev = current
            current = current.next
        raise KeyError(f"Ключ {key} не найден")

    def _resize(self, new_size):
        old_table = self.table
        self.size = self._next_prime(new_size)
        self.table = [UnorderedList() for _ in range(self.size)]
        self.count = 0
        for bucket in old_table:
            current = bucket.head
            while current:
                self.put(current.data[0], current.data[1])
                current = current.next


if __name__ == "__main__":
    ht = HashTableChains()
    for i in range(20):
        ht[i] = i * 10
    print("Размер таблицы:", len(ht))
    print("10 в таблице?", 10 in ht)
    print("Значение по ключу 5:", ht[5])
    del ht[5]
    print("5 в таблице?", 5 in ht)
    print("Размер после удаления:", len(ht))
```

```python
# 4 задание

class HashTable:
    def __init__(self, size=11):
        self.size = self._next_prime(size)
        self.slots = [None] * self.size
        self.data = [None] * self.size
        self.count = 0

    def _next_prime(self, n):
        p = n if n % 2 else n + 1
        while True:
            if all(p % i != 0 for i in range(3, int(p**0.5) + 1, 2)):
                return p
            p += 2

    def _hash(self, key, step=0):
        if isinstance(key, str):
            key = sum(ord(c) for c in key)
        return (key % self.size + step * step) % self.size

    def put(self, key, val):
        if self.count / self.size >= 0.7:
            self._resize(self.size * 2)
        step = 0
        start = self._hash(key)
        pos = start
        while self.slots[pos] not in (None, key):
            step += 1
            pos = self._hash(key, step)
        if self.slots[pos] is None:
            self.count += 1
        self.slots[pos] = key
        self.data[pos] = val

    def get(self, key):
        step = 0
        start = self._hash(key)
        pos = start
        while self.slots[pos] is not None:
            if self.slots[pos] == key:
                return self.data[pos]
            step += 1
            pos = self._hash(key, step)
        return None

    def __getitem__(self, key):
        return self.get(key)

    def __setitem__(self, key, val):
        self.put(key, val)

    def __contains__(self, key):
        return self.get(key) is not None


# подсчёт вхождений слов
def count_words(text):
    ht = HashTable()
    words = text.split()
    result = []

    for word in words:
        if word in ht:
            ht[word] += 1
        else:
            ht[word] = 1
        result.append(str(ht[word]))

    return " ".join(result)


if __name__ == "__main__":
    text = "Раз раз раз как меня слышно Повторяю раз раз раз Повторяю"
    print(count_words(text))
```

```python
# 5 задание

import hashlib
import secrets
import json
import os

USER_FILE = "users.json"

def load_users():
    if os.path.exists(USER_FILE):
        with open(USER_FILE, "r") as f:
            return json.load(f)
    return {}

def save_users(users):
    with open(USER_FILE, "w") as f:
        json.dump(users, f)

def register():
    login = input("Логин: ")
    if login in load_users():
        print("Пользователь уже существует")
        return
    password = input("Пароль: ").encode()
    salt = secrets.token_bytes(32)
    hash_pass = hashlib.scrypt(password, salt=salt, n=16384, r=8, p=1, dklen=32)

    users = load_users()
    users[login] = {
        "salt": salt.hex(),
        "hash": hash_pass.hex()
    }
    save_users(users)
    print("Регистрация успешна")

def login():
    login = input("Логин: ")
    users = load_users()
    if login not in users:
        print("Пользователь не найден")
        return
    password = input("Пароль: ").encode()
    salt = bytes.fromhex(users[login]["salt"])
    hash_pass = hashlib.scrypt(password, salt=salt, n=16384, r=8, p=1, dklen=32)

    if hash_pass.hex() == users[login]["hash"]:
        print("Вход выполнен")
    else:
        print("Неверный пароль")

def main():
    while True:
        print("\n1 - Регистрация\n2 - Вход\n3 - Выход")
        choice = input("Выбор: ")
        if choice == "1":
            register()
        elif choice == "2":
            login()
        elif choice == "3":
            break

if __name__ == "__main__":
    main()
```

```python
# 6 задание

import hashlib
import os
from google.colab import files

# загружаем файлы
print("загрузите файлы дубликаты")
uploaded = files.upload()

hashes = {}

for filename, content in uploaded.items():
    hash_algo = hashlib.sha256()
    hash_algo.update(content)
    file_hash = hash_algo.hexdigest()

    if file_hash not in hashes:
        hashes[file_hash] = []
    hashes[file_hash].append(filename)

print("\n ДУБЛИКАТЫ \n")
found = False
for h, files in hashes.items():
    if len(files) > 1:
        print(f"Хеш {h[:16]}...")
        print(f"Файлы {', '.join(files)}\n")
        found = True

if not found:
    print("Дубликатов не найдено")
```
