```python
# 1 задание
import time
import matplotlib.pyplot as plt

    #1 Преобразует целое неотрицательное число в строку.
    #2 Сложность: O(k), где k — количество цифр числа.

def foo(i):
    digits = "0123456789"
    if i == 0:
        return "0"
    result = ""
    while i > 0:
        result = digits[i % 10] + result
        i //= 10
    return result


ns = [10**k for k in range(1, 7)]
times = []
repeats = 100000

for n in ns:
    start = time.perf_counter()
    for _ in range(repeats):
        foo(n)
    end = time.perf_counter()
    times.append((end - start) / repeats)

plt.figure()
plt.plot(ns, times, 'o-')
plt.xscale('log')
plt.xlabel('Число n')
plt.ylabel('Время (сек)')
plt.title('Зависимость времени от входных данных (среднее на вызов)')
plt.grid(True)
plt.show()
```

```python
# 2 задание (график)

import time
import matplotlib.pyplot as plt

def all_unique_set(lst):
    return len(lst) == len(set(lst))

def all_unique_loops(lst):
    for i in range(len(lst)):
        for j in range(i + 1, len(lst)):
            if lst[i] == lst[j]:
                return False
    return True

sizes = [100, 200, 400, 800, 1600, 3200]
times_set = []
times_loops = []

for size in sizes:
    lst = list(range(size))

    start = time.perf_counter()
    all_unique_set(lst)
    times_set.append(time.perf_counter() - start)

    start = time.perf_counter()
    all_unique_loops(lst)
    times_loops.append(time.perf_counter() - start)

plt.plot(sizes, times_set, 'o-', label='O(n)')
plt.plot(sizes, times_loops, 's-', label='O(n²)')
plt.xlabel('Размер списка')
plt.ylabel('Время (сек)')
plt.legend()
plt.grid()
plt.show()
```

```python
# 4 задание
import time
import matplotlib.pyplot as plt

sizes = [1000, 2000, 4000, 8000, 16000, 32000]
times_dict = []
times_list = []

for size in sizes:
    d = {i: i for i in range(size)}
    lst = list(range(size))

    # del для словаря
    start = time.perf_counter()
    del d[size // 2]
    times_dict.append(time.perf_counter() - start)

    # del для списка
    start = time.perf_counter()
    del lst[size // 2]
    times_list.append(time.perf_counter() - start)

plt.plot(sizes, times_dict, 'o-', label='dict del O(1)')
plt.plot(sizes, times_list, 's-', label='list del O(n)')
plt.xlabel('Размер')
plt.ylabel('Время (сек)')
plt.legend()
plt.grid()
plt.show()
```

```python
# 5 задание
import time
import matplotlib.pyplot as plt

sizes = [1000, 2000, 4000, 8000, 16000, 32000]
times_set = []
times_list = []

for size in sizes:
    s = set(range(size))
    lst = list(range(size))

    # in для множества
    start = time.perf_counter()
    (size - 1) in s
    times_set.append(time.perf_counter() - start)

    # in для списка
    start = time.perf_counter()
    (size - 1) in lst
    times_list.append(time.perf_counter() - start)

plt.plot(sizes, times_set, 'o-', label='set in O(1)')
plt.plot(sizes, times_list, 's-', label='list in O(n)')
plt.xlabel('Размер')
plt.ylabel('Время (сек)')
plt.legend()
plt.grid()
plt.show()
```
