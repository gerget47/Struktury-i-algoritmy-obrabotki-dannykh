````python
# 1 задание

import time
import random
import matplotlib.pyplot as plt

# сортировка выбором
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# быстрая сортировка
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# замер времени
def measure_time(sort_func, data):
    start = time.perf_counter()
    sort_func(data.copy())
    return time.perf_counter() - start

sizes = [100, 200, 400, 800, 1600, 3200]

data_types = {
    'Случайные': lambda size: [random.randint(0, size) for _ in range(size)],
    'Отсортированные': lambda size: list(range(size)),
    'Обратные': lambda size: list(range(size, 0, -1))
}

results_selection = {name: [] for name in data_types}
results_quick = {name: [] for name in data_types}

# измерения
for size in sizes:
    for name, generator in data_types.items():
        data = generator(size)
        results_selection[name].append(measure_time(selection_sort, data))
        results_quick[name].append(measure_time(quick_sort, data))

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, (name, ax) in enumerate(zip(data_types.keys(), axes)):
    ax.plot(sizes, results_selection[name], 'o-', label='Выбором O(n²)')
    ax.plot(sizes, results_quick[name], 's-', label='Быстрая O(n log n)')
    ax.set_xlabel('Размер массива')
    ax.set_ylabel('Время (сек)')
    ax.set_title(name)
    ax.legend()
    ax.grid()

plt.tight_layout()
plt.show()
````

````python
# 2 задание

def tribonacci(n):
# базовые случаи
    if n == 0 or n == 1:
        return 0
    if n == 2:
        return 1
# рекурсивные случаи
    return tribonacci(n - 1) + tribonacci(n - 2) + tribonacci(n - 3)

for i in range(10):
    print(f"trib({i}) = {tribonacci(i)}")
````

````python
# 3 задание

def recursive_count(some_list, t, index=0):

    if index == len(some_list):
        return 0

    if some_list[index] == t:
        return 1 + recursive_count(some_list, t, index + 1)
    else:
        return recursive_count(some_list, t, index + 1)

print(recursive_count([1, 2, 3, 2, 4, 2, 5], 2))
````

````python
# 4 задание

import time
import matplotlib.pyplot as plt

# рекурсивное вычисление Фибоначчи
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

# рекурсивное вычисление Люка
def lucas(n):
    if n == 0:
        return 2
    if n == 1:
        return 1
    return lucas(n-1) + lucas(n-2)

# Фибоначчи через числа Люка
def fib_with_lucas(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    i = n // 2
    j = n - i
    Fi = fib_with_lucas(i)
    Lj = lucas(j)
    Li = lucas(i)
    return (Fi + Lj) * (Fi + Li) // 2

# сравнение скорости
n_values = list(range(10, 31, 2))
times_fib = []
times_fib_lucas = []

for n in n_values:

    start = time.perf_counter()
    fibonacci(n)
    times_fib.append(time.perf_counter() - start)


    start = time.perf_counter()
    fib_with_lucas(n)
    times_fib_lucas.append(time.perf_counter() - start)

plt.plot(n_values, times_fib, 'o-', label='fibonacci (медленный)')
plt.plot(n_values, times_fib_lucas, 's-', label='fib_with_lucas (быстрый)')
plt.xlabel('n')
plt.ylabel('Время (сек)')
plt.legend()
plt.grid()
plt.show()
````

import turtle
import random


def tree(branchLen, t, width):
    if branchLen > 5:
        t.pensize(width)

        if branchLen < 15:
            t.color("green")
        else:
            t.color("brown")

        t.forward(branchLen)

        angle = random.uniform(15, 45)
        t.right(angle)

        new_len = branchLen - random.uniform(10, 20)
        tree(new_len, t, width * 0.7)

        t.left(angle * 2)
        tree(new_len, t, width * 0.7)

        t.right(angle)
        t.backward(branchLen)


def main():
    t = turtle.Turtle()
    myWin = turtle.Screen()
    t.left(90)
    t.up()
    t.backward(100)
    t.down()
    t.speed(0)

main()

import random
import turtle


def mountain(x1, y1, x2, y2, roughness, t):
    if abs(x2 - x1) < 10:
        t.up()
        t.goto(x1, y1)
        t.down()
        t.goto(x2, y2)
    else:
        mid_x = (x1 + x2) / 2
        mid_y = (y1 + y2) / 2
        mid_y += random.uniform(-roughness, roughness)

        mountain(x1, y1, mid_x, mid_y, roughness / 2, t)
        mountain(mid_x, mid_y, x2, y2, roughness / 2, t)


t = turtle.Turtle()
t.speed(0)
t.up()
t.goto(-300, -50)
t.down()

mountain(-300, -50, 300, -50, 100, t)

````python
# 7 задание

def print_matrix(matrix):
    for row in matrix:
        print(''.join(map(str, row)))

def is_valid(matrix, row, col, num):

    for j in range(4):
        if matrix[row][j] == num:
            return False

    for i in range(4):
        if matrix[i][col] == num:
            return False

    start_row, start_col = (row // 2) * 2, (col // 2) * 2
    for i in range(start_row, start_row + 2):
        for j in range(start_col, start_col + 2):
            if matrix[i][j] == num:
                return False
    return True

def solve_sudoku(matrix):
    for i in range(4):
        for j in range(4):
            if matrix[i][j] == 0:
                for num in range(1, 5):
                    if is_valid(matrix, i, j, num):
                        matrix[i][j] = num
                        if solve_sudoku(matrix):
                            return True
                        matrix[i][j] = 0
                return False
    return True


sudoku = [
    [0, 0, 0, 0],
    [0, 0, 2, 0],
    [0, 1, 0, 0],
    [3, 0, 0, 4]
]

print("Исходная:")
print_matrix(sudoku)

if solve_sudoku(sudoku):
    print("\nРешённая:")
    print_matrix(sudoku)
else:
    print("\nРешения нет")
````
