```python
# задание 3

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# лучший случай O(n)
def bubble_sort_optimized(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

# замер времени
import time
import matplotlib.pyplot as plt

sizes = [100, 500, 1000, 2000, 4000, 6000, 8000, 10000]
times_best = []
times_worst = []

for n in sizes:
    # Лучший случай
    arr_best = list(range(n))
    start = time.perf_counter()
    bubble_sort_optimized(arr_best)
    times_best.append(time.perf_counter() - start)

    # Худший случай
    arr_worst = list(range(n, 0, -1))
    start = time.perf_counter()
    bubble_sort(arr_worst)
    times_worst.append(time.perf_counter() - start)

plt.plot(sizes, times_best, 'o-', label='Лучший случай O(n)')
plt.plot(sizes, times_worst, 's-', label='Худший случай O(n²)')
plt.xlabel('Размер списка')
plt.ylabel('Время (сек)')
plt.legend()
plt.grid()
plt.show()
```

```python
# задание 4

import time
import random
import matplotlib.pyplot as plt

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

def bubble_sort_optimized(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

# генераторы данных
def random_list(n):
    return [random.randint(0, 10000) for _ in range(n)]

def sorted_list(n):
    return list(range(n))

def almost_reversed(n):
    arr = list(range(n, 0, -1))
    mid = n // 2
    arr[mid], arr[mid + 1] = arr[mid + 1], arr[mid]
    return arr

def half_zero_half_one(n):
    return [0] * (n // 2) + [1] * (n - n // 2)

def scattered(n):
    arr = list(range(n))
    for _ in range(int(n * 0.05)):
        i, j = random.randint(0, n-1), random.randint(0, n-1)
        arr[i], arr[j] = arr[j], arr[i]
    return arr

sizes = [100, 500, 1000, 2000, 4000, 6000, 8000, 10000]
generators = [
    ("Случайные", random_list, bubble_sort),
    ("Отсортированные", sorted_list, bubble_sort_optimized), # самая быстрая из-за оптимизации
    ("Почти обратные", almost_reversed, bubble_sort),  # очень медленно, так как почти все элементы не на своих местах, из-за этог требуется много перестановок
    ("0 и 1", half_zero_half_one, bubble_sort),
    ("5% разброс", scattered, bubble_sort_optimized)
]

plt.figure(figsize=(12, 10))
for i, (name, gen, sort_func) in enumerate(generators, 1):
    times = []
    for n in sizes:
        arr = gen(n)
        start = time.perf_counter()
        sort_func(arr.copy())
        times.append(time.perf_counter() - start)

    plt.subplot(2, 3, i)
    plt.plot(sizes, times, 'o-')
    plt.title(name)
    plt.xlabel('Размер')
    plt.ylabel('Время (сек)')
    plt.grid()

plt.tight_layout()
plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt

# обычная сортировка пузырьком
def bubble_sort_simple(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# модифицированная сортировка
def bubble_sort_modified(arr):
    n = len(arr)
    left = 0
    right = n - 1
    while left < right:
        last_swap = left
        for j in range(left, right):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                last_swap = j
        right = last_swap

        if left >= right:
            break

        last_swap = right
        for j in range(right, left, -1):
            if arr[j] < arr[j - 1]:
                arr[j], arr[j - 1] = arr[j - 1], arr[j]
                last_swap = j
        left = last_swap


# генераторы данных
def random_list(n):
    return [random.randint(0, 10000) for _ in range(n)]

def sorted_list(n):
    return list(range(n))

def almost_reversed(n):
    arr = list(range(n, 0, -1))
    mid = n // 2
    arr[mid], arr[mid + 1] = arr[mid + 1], arr[mid]
    return arr

def half_zero_half_one(n):
    return [0] * (n // 2) + [1] * (n - n // 2)

def scattered(n):
    arr = list(range(n))
    for _ in range(int(n * 0.05)):
        i, j = random.randint(0, n-1), random.randint(0, n-1)
        arr[i], arr[j] = arr[j], arr[i]
    return arr


# размеры и типы данных
sizes = [100, 500, 1000, 2000, 4000, 6000, 8000, 10000]
data_types = [
    ("Случайные", random_list),
    ("Отсортированные", sorted_list),
    ("Почти обратные", almost_reversed),
    ("0 и 1", half_zero_half_one),
    ("5% разброс", scattered)
]

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for idx, (name, generator) in enumerate(data_types):
    times_simple = []
    times_modified = []

    for n in sizes:
        arr = generator(n)

        arr1 = arr.copy()
        start = time.perf_counter()
        bubble_sort_simple(arr1)
        times_simple.append(time.perf_counter() - start)

        arr2 = arr.copy()
        start = time.perf_counter()
        bubble_sort_modified(arr2)
        times_modified.append(time.perf_counter() - start)

    axes[idx].plot(sizes, times_simple, 'o-', label='Обычная')
    axes[idx].plot(sizes, times_modified, 's-', label='Модифицированная')
    axes[idx].set_title(name)
    axes[idx].set_xlabel('Размер списка')
    axes[idx].set_ylabel('Время (сек)')
    axes[idx].legend()
    axes[idx].grid()

plt.tight_layout()
plt.show()


print("Сравнение на случайных данных (n=2000):")
arr = random_list(2000)

arr1 = arr.copy()
start = time.perf_counter()
bubble_sort_simple(arr1)
t_simple = time.perf_counter() - start

arr2 = arr.copy()
start = time.perf_counter()
bubble_sort_modified(arr2)
t_modified = time.perf_counter() - start

print(f"Обычная сортировка: {t_simple:.4f} сек")
print(f"Модифицированная: {t_modified:.4f} сек")
print(f"Ускорение: {t_simple/t_modified:.2f}x")

print("\nсравнение на почти отсортированных данных (n=2000)")
arr = almost_reversed(2000)

arr1 = arr.copy()
start = time.perf_counter()
bubble_sort_simple(arr1)
t_simple = time.perf_counter() - start

arr2 = arr.copy()
start = time.perf_counter()
bubble_sort_modified(arr2)
t_modified = time.perf_counter() - start

print(f"обычная сортировка: {t_simple:.4f} сек")
print(f"модифицированная: {t_modified:.4f} сек")
print(f"ускорение: {t_simple/t_modified:.2f}x")
```
